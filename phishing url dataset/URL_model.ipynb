```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import numpy as np
```

```python
df_train = pd.read_parquet("Training.parquet")
df_test = pd.read_parquet("Testing.parquet")
df_train
```

```python
clean_train = df_train.copy()
clean_test = df_test.copy()
```

```python
clean_train = clean_train.drop(columns=["url"])
clean_test = clean_test.drop(columns=["url"])
```

```python
# 1 = legitimate
# 2 = phishing
clean_train["status"] = clean_train["status"].replace({"legitimate":1, "phishing":0})
clean_test["status"] = clean_test["status"].replace({"legitimate":1, "phishing":0})
```

```python
clean_train
```

```python
y = clean_train["status"].values
X = clean_train.drop(columns="status").values

yy = clean_test["status"].values
XX = clean_test.drop(columns="status").values

y_train = y
X_train = X

y_test = yy
X_test = XX
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
